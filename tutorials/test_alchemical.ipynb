```python
s = df["Composite_rule_sizes"].astype(str).str.strip()

# rows with multiple values, e.g. "2,3"
is_mixed = s.str.contains(",", na=False)

# counts for single values only
single_counts = s[~is_mixed].value_counts().sort_index()

# add mixed count
stats = pd.concat([
    single_counts,
    pd.Series({"mixed": is_mixed.sum()})
])

stats
```
