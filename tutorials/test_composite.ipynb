```python
import json
import pandas as pd
from argparse import Namespace
from pathlib import Path

from alchems.composite_rules.extract import (
    SynPlannerRuleExtractor,
    RouteProcessingStats,
    normalize_route_tree,
    extract_route_composites,
)

routes_path = Path("/Users/almazgil/Desktop/projects/Retro-BLEU/PaRoutes/data/n1-routes.json")
routes = json.load(routes_path.open())
routes_json = {i: route for i, route in enumerate(routes)}

route_id = 17
route = normalize_route_tree(routes_json[route_id])

args = Namespace(
    config=Path("/Users/almazgil/Desktop/projects/Retro-BLEU/composite_rules/configs/rule_extraction_functional_groups.yaml"),
    environment_atom_count=1,
    include_rings=False,
    keep_leaving_groups=True,
    keep_incoming_groups=False,
    reactor_validation=False,
)

extractor = SynPlannerRuleExtractor.from_args(args)
stats = RouteProcessingStats()

comp_rule_res = extract_route_composites(
    route,
    extractor,
    min_length=2,
    max_length=5,
    stats=stats,
)

rows = []
for seq, target_molecules in sorted(
    comp_rule_res.items(),
    key=lambda item: (len(item[0]), "$".join(item[0])),
):
    rows.append({
        "size": len(seq),
        "Composite_rule": "$".join(seq),
        "Target_molecules": ",".join(sorted(target_molecules)),
        **{f"t{i+1}": rule for i, rule in enumerate(seq)},
    })

route_df = pd.DataFrame(rows)
display(route_df)
print(stats)
```
